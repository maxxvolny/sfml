```python
#средний возраст женщин
sum_age = 0

for j in range(0,len(w_age)):
    if wl[j] == 2:
        sum_age += w_age[j]
           
            

sum_age /= wc
print("Mean woman age %.2f" % (sum_age))

```

```python
#средний возраст мужчин
sum_age = 0

for j in range(0,len(w_age)):
    if wl[j] == 1:
        sum_age += w_age[j]
           
            
sum_age /= mc
print("Mean man age %.2f" % (sum_age))
```

```python
#средний возраст женатых мужчин
sum_age = 0
mc = 0

for j in range(0,len(w_age)):
    if (wl[j] == 1) & (w_married[j] == 2):
        sum_age += w_age[j]
        mc += 1
           
if mc > 0:
    sum_age /= mc
else:
    sum_age = 0

print("Mean married man age %.2f" % (sum_age))
```

```python
#средний возраст не женатых мужчин
sum_age = 0
mc = 0

for j in range(0,len(w_age)):
    if (wl[j] == 1) & (w_married[j] == 1):
        sum_age += w_age[j]
        mc += 1
           
if mc > 0:
    sum_age /= mc
else:
    sum_age = 0
    
    
print("Mean unmarried man age %.2f" % (sum_age))
```

```python
# средний предел баланса не женатых мужчин
sum_bal = 0
mc = 0


for j in range(0,len(w_age)):
    if (wl[j] == 1) & (w_married[j] == 1):
        sum_bal += w_LimBal[j]
        mc += 1
           
if mc > 0:
    sum_bal /= mc
else:
    sum_bal = 0
    
print("Mean unmarried man balance limit %.3f" % (sum_bal))

```

```python
# медиана средней суммы платежей замужних женщин

med_pay = list();


for j in range(0,len(wl)):
    if (wl[j] == 2) & (w_married[j] == 2):
        med_pay.append(w_WSP[j])
        
        af =  pd.Series(med_pay)
      
    
print("Median unmarried woman payment %.2f" % (af.median()))

```

```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

#TODO
```

```python
new_l = list()

for j in range(0,len(w_LimBal)):
    if w_LimBal[j] <=10000:
        new_l.append('A')
    elif (w_LimBal[j] >10000) & (w_LimBal[j] <=100000):
        new_l.append('B')
    elif (w_LimBal[j] >100000) & (w_LimBal[j] <=200000):
        new_l.append('C')   
    elif (w_LimBal[j] >200000) & (w_LimBal[j] <=400000):
        new_l.append('D')       
    elif (w_LimBal[j] >400000) & (w_LimBal[j] <=700000):
        new_l.append('E')
    elif w_LimBal[j] >700000:
        new_l.append('F')
 
df = df.copy()
df['NEW_ABCDEF'] = pd.Series(new_l)
df.head(10)
```
